```python
import json
import os
from IPython.display import display, HTML
import pandas as pd
import numpy as np

result_dir = "Results/"

simulations_names = os.listdir(result_dir)
simulations = []
for name in simulations_names:
    if os.path.isfile(result_dir + name):
        continue
    with open(result_dir + name + "/results.json") as f:
        simulations.append(json.load(f))
table = []
for simulation in simulations:
    table.append([simulation["model"], simulation["lr"], simulation["kernel"], str(simulation["filters"]), simulation["bn"], simulation["wn"],
                simulation["metrics"]["test_metrics"]["mse"], simulation["metrics"]["test_metrics"]["ux"], simulation["metrics"]["test_metrics"]["uy"],
                simulation["metrics"]["test_metrics"]["p"]])
df = pd.DataFrame(table, columns = ["Name","Lr","Kernel","Filters","BN","WN","Mse","Ux","Uy","P"])
df
```

```python
print("Sorting by MSE:")
df.sort_values("Mse")
```

```python
print("Sorting by Ux:")
df.sort_values("Ux")
```

```python
print("Sorting by Uy:")
df.sort_values("Uy")
```

```python
print("Sorting by p:")
df.sort_values("P")
```
